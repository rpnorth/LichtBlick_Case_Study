# Running the final version for the Analytics Engineer Case Study

```python
from load_date_create_table import CreateRevenueTable
import pandas as pd
```

## Run the ETL process for each set of data (based on download date)

```python
d_2020_10 = CreateRevenueTable(period_start_date = '20190101', download_date = '20201001')
d_2020_10.run_load_data()
d_2020_10.run_create_revenue_table()
display(d_2020_10.df_output)

d_2020_12 = CreateRevenueTable(period_start_date = '20201002', download_date = '20201201')
d_2020_12.run_load_data()
d_2020_12.run_create_revenue_table()
display(d_2020_12.df_output)

d_2021_01 = CreateRevenueTable(period_start_date = '20201202', download_date = '20210101')
d_2021_01.run_load_data()
d_2021_01.run_create_revenue_table()
display(d_2021_01.df_output)


```

## Combine the three resulting tables for upload to the data warehouse
- the download_date is included to allow a user to easily separate the results and get a history

```python
df_for_DWH = pd.concat([d_2020_10.df_output, d_2020_12.df_output, d_2021_01.df_output]).sort_values('customer_id')
df_for_DWH
```

## Calculate the average revenue per contract for three different time periods
- this info will be used when presenting to the stakeholder

```python
# average revenue per contract
calc_vals = {'download_date': [pd.to_datetime('20201001').normalize(), pd.to_datetime('20201201').normalize(), pd.to_datetime('20210101').normalize()],
            'number_of_contracts': [d_2020_10.number_contracts, d_2020_12.number_contracts, d_2021_01.number_contracts], 
            'total_revenue': [d_2020_10.total_revenue, d_2020_12.total_revenue, d_2021_01.total_revenue], 
            'average_revenue_per_contract': [d_2020_10.av_revenue_per_contract, d_2020_12.av_revenue_per_contract, d_2021_01.av_revenue_per_contract],
            'total_average_annual_revenue': [d_2020_10.total_annual_revenue, d_2020_12.total_annual_revenue, d_2021_01.total_annual_revenue], 
            'average_annual_revenue_per_contract': [d_2020_10.annual_av_revenue_per_contract, d_2020_12.annual_av_revenue_per_contract, d_2021_01.annual_av_revenue_per_contract]}

df_calc_vals = pd.DataFrame.from_dict(calc_vals, orient='index')
df_calc_vals.columns = ['2019-01 to 2020-10', '2020-10 to 2020-12', '2020-12 to 2021-01']
df_calc_vals
```
